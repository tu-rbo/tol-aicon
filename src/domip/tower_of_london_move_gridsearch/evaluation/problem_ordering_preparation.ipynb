```python
from pathlib import Path
import pandas as pd

# Resolve paths relative to this notebook location.
notebook_dir = Path.cwd()
base_dir = notebook_dir.parents[2] / "tower_of_london_human_data"
input_csv = base_dir / "combined_processed.csv"
output_dir = base_dir / "problems_ordered"
output_dir.mkdir(parents=True, exist_ok=True)

# Load source data.
df = pd.read_csv(input_csv)

labels = ["Healthy", "PD", "MCI", "Stroke"]

# Create one output per metric column and label.
# Exclude non-ordering columns and keep numeric metrics.
ordering_columns = [
    c
    for c in df.columns
    if c not in {"problem", "label"} and pd.api.types.is_numeric_dtype(df[c])
]

for column_name in ordering_columns:
    for label in labels:
        ordered = (
            df.loc[df["label"] == label, ["problem", column_name]]
            .sort_values(column_name, ascending=True)
            .reset_index(drop=True)
        )

        # Convert problem 0..23 to TOL-ID-1..TOL-ID-24.
        ordered["TOL-ID"] = ordered["problem"].astype(int).add(1).map(lambda x: f"TOL-ID-{x}")

        # Keep only required columns in required order.
        result = ordered[["TOL-ID", column_name]]

        out_file = output_dir / f"{column_name}_{label}.csv"
        result.to_csv(out_file, index=False)

        print(f"{label} | {column_name}: wrote {len(result)} rows to {out_file}")

# Quick verification table.
summary = {
    f"{column}_{label}": len(pd.read_csv(output_dir / f"{column}_{label}.csv"))
    for column in ordering_columns
    for label in labels
}
summary
```
